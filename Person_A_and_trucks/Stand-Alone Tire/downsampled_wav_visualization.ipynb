```python
# import libraries
import os
import numpy as np
import soundfile as sf
import plotly.graph_objects as go
import plotly.express as px


'''
This script is used to plot the waveform of a selected downsampled WAV audio file from /Users/jothamwambi/Projects/tire_analysis/Stand-Alone Tire/downsampled/15_225-80R17.5-700-710R_H.wav using pydub and plotly
'''

# load the audio file
audio_file = 'downsampled_2K/01_11R22.5-900-710R_H.wav'
audio_data, sample_rate = sf.read(audio_file)

# plot the waveform
fig = go.Figure()
fig.add_trace(go.Scatter(x=np.arange(len(audio_data))/sample_rate, y=audio_data, mode='lines'))
fig.update_layout(title='Waveform of the audio file', xaxis_title='Time (s)', yaxis_title='Amplitude')
fig.show()


```

```python
# Save the plot as an HTML file to the root directory with the same name as the audio file
html_file = os.path.splitext(os.path.basename(audio_file))[0] + '.html'
fig.write_html(html_file)
print(f'Waveform plot saved as {html_file}')
```

```python
# Play selected audio file using pydub with a play button to play the audio
import pydub
from pydub import AudioSegment
from pydub.playback import play
import ipywidgets as widgets
from IPython.display import display

# load the audio file
audio = AudioSegment.from_file(audio_file)

# play the audio
#play(audio)



# create a button
button = widgets.Button(description="Play Audio")

# define the function to play audio on button click
def on_button_click(b):
    play(audio)

# bind the function to the button click event
button.on_click(on_button_click)

# display the button
display(button)
```
